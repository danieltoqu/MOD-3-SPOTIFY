# Data Analysis - Linear Regression

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Creating artist_codes dataframe
artists = pd.read_csv('Data/spotify_artists_cleaned.csv')
```

```python
df = pd.DataFrame(artists, columns=['artist_popularity', 'artist_followers'])
```

```python
print(df.info())
df.head()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, tight_layout=True)

ax1.boxplot(df.artist_popularity)
ax1.set_title('Popularity')
ax2.boxplot(df.artist_followers)
ax2.set_title('Followers');
```

```python
sns.violinplot(data=df.artist_popularity)
```

```python
sns.violinplot(data=df.artist_followers)
```

```python
plt.figure(figsize= (10,8))
plt.scatter(x = df.artist_popularity, y = df.artist_followers)
plt.xlabel('popularity')
plt.ylabel('followers')
plt.show()
```

```python
df.quantile(q=0.98)
```

```python
idx = (df < df.quantile(q=0.98)).all(axis=1)
df2 = df[idx]
```

```python
df2.info()
```

```python
sns.violinplot(data=df2.artist_popularity)
```

```python
sns.violinplot(data=df2.artist_followers)
```

```python
plt.figure(figsize= (10,8))
plt.scatter(x = df2.artist_popularity, y = df2.artist_followers)
plt.xlabel('popularity')
plt.ylabel('followers')
plt.show()
```
